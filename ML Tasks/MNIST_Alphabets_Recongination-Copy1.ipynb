```python
import pandas as pd
import zipfile
import os
import cv2
```

```python
def read_csv_from_zip(zip_path, csv_filename):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        with zip_ref.open(csv_filename) as file:
            df = pd.read_csv(file)
    return df
```

```python
alpha = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
arr = []
label = [] 
for letter in alpha:
   
    os.chdir(f'E:\\#Data Science Innomatics\\Module 6 Ai-ML\\Datasets\\New folder\\{letter}')
    
    for i in os.listdir():
        path = os.getcwd()+f'\\{i}'
        image = cv2.imread(path)
        image = cv2.resize(image,(5,4))
        arr.append(image.flatten())
        label.append(letter) 
```
