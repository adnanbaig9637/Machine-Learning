```python
arr = []
label = []
for i  in os.listdir('D:\\innomatics\\ML\\Bone_fracture\\train\\fractured'):
    path = os.getcwd()+f'\\{i}'
    
    image = cv2.imread(path)
    if image is not None:
        im = cv2.resize(image,(12,10))
        arr.append(im.flatten())
        label.append('fractured')
    
```

```python
fractured_df = pd.DataFrame(arr)
fractured_df['Label'] = label
```

```python
fractured_df.shape
```

```python
os.listdir(r'D:\\innomatics\\ML\\Bone_fracture\\train\\not fractured')
```

```python
arr = []
label = []
for i  in os.listdir('D:\\innomatics\\ML\\Bone_fracture\\train\\not fractured'):
    path = os.getcwd()+f'\\{i}'
    
    image = cv2.imread(path)
    if image is not None:
        im = cv2.resize(image,(12,10))
        arr.append(im.flatten())
        label.append('not fractured')
```

```python
not_fractured = pd.DataFrame(arr)
not_fractured['Label'] = label
```

```python
not_fractured
```

```python
not_fractured.shape
```

```python
df = pd.concat([fractured_df,not_fractured],axis=0)
```

```python
df.head()
```

```python
df['Label'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
X = df.drop('Label',axis=1)
y = df['Label']
```

```python
X_s,y_s = smote.fit_resample(X,y)
```

```python
y_s.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_s,y_s,test_size=0.2,random_state=23)
```

# logistic Regression

```python


grid = GridSearchCV(
    estimator=LogisticRegression(),
    param_grid={
        'C': [0.1, 1, 10],
        'penalty': ['l2']
    },
    cv=5,
    return_train_score=True
)

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.score(X_test,y_test)
```

# Decession tree classifier

```python
grid_dt = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid={
        'max_depth': [None, 5, 10]
    },
    cv=5,
    return_train_score=True
)

grid_dt.fit(X_train, y_train)
```

```python
grid_dt.best_params_
```

```python
grid_dt.best_score_
```

```python
grid_dt.score(X_test,y_test)
```

# Random forest classiffier

```python
grid_rf = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid={
        'n_estimators': [50, 100, 200]
    },
    cv=5,
    return_train_score=True
)

grid_rf.fit(X_train, y_train)
```

```python
grid_rf.best_params_
```

```python
grid_rf.best_score_
```

```python
grid_rf.score(X_test,y_test)
```
