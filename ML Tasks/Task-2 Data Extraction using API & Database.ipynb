### Data Extraction using API

#### Data Extraction of SCHOOLS IN HYDERABAD using RapidAPI

```python
import requests
import pandas as pd

url = "https://maps-data.p.rapidapi.com/searchmaps.php"

querystring = {"query":"school","limit":"200","country":"us","lang":"en","lat":"17.4065","lng":"78.4772","offset":"0","zoom":"13"}

headers = {
	"X-RapidAPI-Key": "4548604807mshbdd27b9dc86a619p1db787jsn527a99548f8b",
	"X-RapidAPI-Host": "maps-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
pd.json_normalize(response.json(), record_path = ['data'])
```
