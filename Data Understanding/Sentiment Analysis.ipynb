### Sentiment Analysis

```python
tags = nit.get_tweets('RohitSharma' , mode = 'user' , number = 50)
```

```python
rohit = pd.json_normalize(tags , record_path = ['tweets'])
rohit
```

```python
rohit['text']
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sent = SentimentIntensityAnalyzer()
```

```python
def polarity(n):
    return sent.polarity_scores(n)['compound']
```

```python
rohit['polarity'] = rohit['text'].apply(polarity)
```

```python
rohit['polarity']
```
