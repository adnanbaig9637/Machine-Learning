```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.ensemble import VotingClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
X , y = make_classification(n_samples=1000 , n_features=4 , class_sep= 0.2)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.2 , random_state=23)
```

**Individually Models**

```python
#Logistic

log = LogisticRegression()
log.fit(X_train , y_train)
y_pred = log.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

```python
#KNN

knn = KNeighborsClassifier()
knn.fit(X_train , y_train)
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

```python
#DT

dt = DecisionTreeClassifier()
dt.fit(X_train , y_train)
y_pred = dt.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

```python
#SVM

svc = SVC()
svc.fit(X_train , y_train)
y_pred = svc.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

**Ensembles Techniques on classification problem**
* **1. Voting Classifier**

```python
from sklearn.ensemble import VotingClassifier
```

```python
models = [('LOG',log),('KNN',knn),('DT',dt),('SVC',svc)]

voting = VotingClassifier(estimators=models)
voting.fit(X_train , y_train)

y_pred = voting.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

**2. Bagging CLassifier**

```python
from sklearn.ensemble import BaggingClassifier
```

```python
#estimater give algorithm
# n_esti how many time algorithm will run
bag = BaggingClassifier(estimator=dt , n_estimators=10 , verbose=2)
bag.fit(X_train , y_train)
```

```python
y_pred = bag.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

```python
bag = BaggingClassifier(estimator=svc , n_estimators=15 , verbose=2)
bag.fit(X_train , y_train)
```

```python
y_pred = bag.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```

**Class Task on Regression problem using Ensembles techniques**

```python
from sklearn.datasets import make_regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
X , y = make_regression(n_samples=1000 , n_features=4 , bias=2)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.2 , random_state=23)
```

```python
#Linear

lin = LinearRegression()
lin.fit(X_train , y_train)
y_pred = lin.predict(X_test)

metrics.mean_squared_error(y_test , y_pred)
```

```python
#KNN

knn = KNeighborsRegressor()
knn.fit(X_train , y_train)
y_pred = knn.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

```python
#DT

dt = DecisionTreeRegressor()
dt.fit(X_train , y_train)
y_pred = dt.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

```python
#SVM

svr = SVR()
svr.fit(X_train , y_train)
y_pred = svr.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

**Voting Regression**

```python
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import BaggingRegressor
```

```python
models = [('LOG',lin),('KNN',knn),('DT',dt),('SVC',svr)]

voting = VotingRegressor(estimators=models)
voting.fit(X_train , y_train)

y_pred = voting.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

**Bagging Regression**

```python
bag = BaggingRegressor(estimator=lin , n_estimators=10 ,verbose=2)
bag.fit(X_train , y_train)

y_pred = bag.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

```python
bag = BaggingRegressor(estimator=svr , n_estimators=10 ,verbose=2)
bag.fit(X_train , y_train)

y_pred = bag.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

```python
bag = BaggingRegressor(estimator=dt , n_estimators=20 ,verbose=2)
bag.fit(X_train , y_train)

y_pred = bag.predict(X_test)
metrics.mean_squared_error(y_test , y_pred)
```

**3. Staking**

```python
from sklearn.ensemble import StackingClassifier , StackingRegressor
```

```python
#final est---knn
#cv-cross validation
#verbose=2--you will see how training happening
estimator = [('LOG',log),('KNN',knn),('DT',dt),('SVC',svc)]
stack = StackingClassifier(estimators=estimator , 
                           final_estimator=knn , 
                           cv=10 , verbose=2)
```

```python
stack.fit(X_train , y_train)
```

```python
y_pred = stack.predict(X_test)
metrics.accuracy_score(y_test , y_pred)
```
