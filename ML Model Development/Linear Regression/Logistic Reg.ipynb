### Regularization
* **Regularization:-** To solve the the problem of overfit
* **How:-** Adding penalty term to the cost function
* **Cost function + Penalty = Regularization Term**

* **Types of Regulaization:-**
    * In Linear Model
        * L1 Regularization(Lasso Regression):- 
            * Cost function + L1
            * arg.min(Sum log(1+e -yiwTxi) + lambda |Wi|
        * L2 Regularization (Ridge Regression):-
            * arg.min(Sum log(1+e -yiwTxi) + lambda (wi)**2
        * L1&L2 Regularization (Elastic Net Regression):-
            * arg.min(Sum log(1+e -yiwTxi) + lambda |Wi| * lambda (wi)**2
    * Decision Tress
    * Neural Network

```python
from sklearn.datasets import make_classification , make_circles , make_biclusters , make_blobs
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , ConfusionMatrixDisplay ,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_classification(n_samples=1000 , n_features=2 , n_redundant=0 , class_sep=4)
```

```python
X = pd.DataFrame(X,columns=['X1' , 'X2'])
#y = pd.DataFrame(y,columns=['y'])
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=23)
```

```python
log = LogisticRegression(penalty=None)
log.fit(X_train , y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
accuracy_score(y_test , y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test , y_pred)).plot()
```
