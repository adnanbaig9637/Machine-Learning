```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\User\Downloads\Salary_dataset_1.csv")
df
```

```python
df.drop('Unnamed: 0' , axis=1 , inplace=True)
```

```python
df
```

```python
X = df['YearsExperience']
Y = df['Salary']
```

```python
sns.scatterplot(data = df , x = 'Salary' , y = 'YearsExperience')
```

**Gradient Desc Alg Method**

```python
m = 0  #Starting slope valie
c = 0   #starting intercept value
L = 0.001  #Learning rate
iteration = 1000   #
n = len(X)
```

```python
for i in range(iteration):
    Y_pred = m * X  + c
    
    D_m = (-2/n) * sum(X * (Y - Y_pred))   
    D_c = (-2/n) * sum((Y - Y_pred)) 
    
    m = m - L * D_m
    c = c - L * D_c
    
print('The minimun loss function at ')
print("The slope is",m)
print("The Intercept is ",c)
```

```python
Y_pred = m *X + c
plt.scatter(X,Y)
plt.scatter(X,Y_pred)
```
