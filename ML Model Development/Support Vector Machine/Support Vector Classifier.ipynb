```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs , make_circles
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.svm import SVC
from sklearn import metrics
from mlxtend import plotting
```

```python
X , y = make_blobs(n_samples=1000,
    n_features=2,
    cluster_std=0.8,
    random_state=3 , centers=2)
```

```python
X  #--inputs features
```

```python
y   #--Output feature
```

```python
sns.scatterplot(x = X[:,0] , y = X[:,1] , hue = y)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=23)
```

**Perfectly classified Data**

**Hyperparameter Kernal == 'linear' --------y=mx+c**

```python
svc = SVC(kernel='linear')
svc.fit(X_train , y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
metrics.accuracy_score(y_test , y_pred)
```

**Hyperparameter Kernal == 'RBF'---------y=e rest to the power -x square**

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train , y_train)

y_pred = rbf.predict(X_test)

metrics.accuracy_score(y_test , y_pred)
```

**Hyperparameter Kernal == 'Polinomial' ------y=m1x degree + m1x + c**

```python
poly = SVC(kernel='poly')
poly.fit(X_train , y_train)

y_pred = poly.predict(X_test)

metrics.accuracy_score(y_test , y_pred)
```

**Hyperparameter Kernal == 'Sigmoid' ------ y= 1/1+e rest to the power -x**

```python
sig = SVC(kernel='poly')
sig.fit(X_train , y_train)

y_pred = sig.predict(X_test)

metrics.accuracy_score(y_test , y_pred)
```

```python
X , y = make_circles(n_samples=2000,
                    noise=0.2,
                    random_state=23,
                    factor=0)
```

```python
#Hyperparameter Kernal == 'RBF'

rbf = SVC(kernel='rbf')
rbf.fit(X_train , y_train)

y_pred = rbf.predict(X_test)

metrics.accuracy_score(y_test , y_pred)
plotting.plot_decision_regions(X = X_train , y = y_train , clf = rbf)
```

```python
#Hyperparameter Kernal == 'Sigmoid'

sig = SVC(kernel='poly')
sig.fit(X_train , y_train)

y_pred = sig.predict(X_test)

metrics.accuracy_score(y_test , y_pred)

plotting.plot_decision_regions(X = X_train , y = y_train , clf = sig)
```

**RBF parameter "Gamma"**

```python
X , y = make_circles(n_samples=2000,
                    noise=0.4,
                    random_state=23,
                    factor=0)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=23)
```

```python
sns.scatterplot(x = X[:,0] , y = X[:,1] , hue = y)
```

```python
#Hyperparameter Kernal == 'RBF'
#Before gamma parameter

rbf = SVC(kernel='rbf')
rbf.fit(X_train , y_train)

y_pred = rbf.predict(X_test)

print(metrics.accuracy_score(y_test , y_pred))
#plotting.plot_decision_regions(X = X_train , y = y_train , clf = rbf)
```

```python
#Hyperparameter Kernal == 'RBF'
#After gamma parameter

rbf = SVC(kernel='rbf' , gamma=2)
rbf.fit(X_train , y_train)

y_pred = rbf.predict(X_test)

print(metrics.accuracy_score(y_test , y_pred))
#plotting.plot_decision_regions(X = X_train , y = y_train , clf = rbf)
```

**gamma is Increase |^|  Accuracy will Increase |^|**
