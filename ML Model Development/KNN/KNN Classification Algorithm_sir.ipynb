**Split the data**

```python
X = df.drop('survived' , axis=1)
y = df['survived']
```

```python
X[['pclass', 'sibsp', 'parch']] = X[['pclass', 'sibsp', 'parch']].astype('object')
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size=0.2, random_state=43)
```

**Data Preprocessing**

**Missing value**

```python
X_train.isna().sum()
```

```python
X_train.dtypes
```

**Seprating categoric and numeric colms**

```python
cat_col = X_train.select_dtypes(include= ['object' , 'category' , 'bool']).columns
```

```python
num_col = X_train.select_dtypes(exclude= ['object' , 'category' , 'bool']).columns
```

```python
cat_col
```

```python
num_col
```

**Missing value treatment , Encoding , Scaling**

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder , LabelEncoder , OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

* **X_train preprocessing**
* **Categorical imputation and encoding**

```python
cat_imp = SimpleImputer(missing_values=np.nan , strategy='most_frequent')
num_imp = SimpleImputer(missing_values=np.nan , strategy='median')
ohe = OneHotEncoder()
std = StandardScaler()
```

```python
cat_imp.fit(X_train[cat_col])
X_train_impute = pd.DataFrame(cat_imp.transform(X_train[cat_col]), columns = cat_imp.get_feature_names_out())
```

```python
ohe.fit(X_train_impute)
X_cat =  pd.DataFrame(ohe.transform(X_train_impute).toarray() , columns=ohe.get_feature_names_out())
```

**Numerical Imputation and Scaling**

```python
num_imp.fit(X_train[num_col])
X_1 = pd.DataFrame(num_imp.transform(X_train[num_col]) , columns=num_imp.get_feature_names_out()) 
std.fit(X_1)
X_num = pd.DataFrame(std.transform(X_1) , columns=std.get_feature_names_out())
```

```python
X_num
```

```python
X_cat
```

```python
X_train = pd.concat([X_num ,X_cat] , axis=1)
```

```python
X_train
```

**X_test preprocessing**

```python
X_test_impute = pd.DataFrame(cat_imp.transform( X_test[cat_col]),columns =cat_imp.get_feature_names_out())
X_cat = pd.DataFrame(ohe.transform(X_test_impute).toarray(),columns = ohe.get_feature_names_out())
X_1 = pd.DataFrame(num_imp.transform(X_test[num_col]),columns = num_imp.get_feature_names_out())
X_num = pd.DataFrame(std.transform(X_1),columns = std.get_feature_names_out())
X_test = pd.concat([X_num,X_cat],axis = 1)
```

```python
X_test
```

**Model Building**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train , y_train)
```

```python
y_pred = knn.predict(X_test.values)
```

```python
y_pred
```

**Model evaluation**

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
confusion_matrix(y_test , y_pred)
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.metrics import precision_score , recall_score
```

```python
precision_score(y_test , y_pred)
```

```python
recall_score(y_test , y_pred)
```
