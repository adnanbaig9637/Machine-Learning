**IMPORTING REQUIRED LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.datasets import make_classification , make_circles , make_blobs , make_biclusters
from mlxtend.plotting import plot_decision_regions
import warnings
warnings.filterwarnings("ignore")
```

**MAKE CLASSIFICATION**

```python
X , y = make_classification(n_samples=2000 , n_features=2 , n_redundant=0 , class_sep=3)
```

```python
X = pd.DataFrame(X , columns=['X1' , 'X2'])
```

**SPLITTING THE DATA**

```python
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size=0.2 , random_state=23)
```

**ALGORITHMS**

```python
log = LogisticRegression(penalty=None)
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
nb = GaussianNB()
```

```python
log.fit(X_train , y_train)
knn.fit(X_train , y_train)
dt.fit(X_train , y_train)
nb.fit(X_train , y_train)
```

```python
y_predict = log.predict(X_test)
y_predict
```

```python
confusion_matrix(y_test , y_predict)
```

```python
accuracy_score(y_test , y_predict)
```

**DECISION BOUNDARIES**

```python
sns.scatterplot(data = X , x = X['X1'] , y = X['X2'] , hue = y)
```

**DECISION REGIONS**

```python
plot_decision_regions(X = X_train.values , y = y_train , clf = log)
```

```python
plot_decision_regions(X = X_train.values , y = y_train , clf = knn)
```

```python
plot_decision_regions(X = X_train.values , y = y_train , clf = dt)
```

```python
plot_decision_regions(X = X_train.values , y = y_train , clf = nb)
```
