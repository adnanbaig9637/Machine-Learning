### Scaling :- Certain Range
#### To reduce influence of larger values (Scale Free)
#### To get unitless

* Standard scalar-- u=0 ,, standDev=1 -- Z= X-u/standDev --approxmately follow normal distribution -- slightly afftected by outliers.
* Minmax Scalar :- Xi = Xi-min(x) / Max(x)-Min(x) -- range min&max = (0,1) -- Not following normal distrubution -- Highly affected by outliers.
* Robust Scalar :- Xi-Xmedian / IQR -- (25th to 75th percentile data) -- normal -- when the data contain marginal outliers.
* Maxabs Scalar :- Xi / |Max(Xi)| -- affected by outliers

**Function Transformer:--**
* Square root
* Square
* Recipocal
* Expponential
* Log

```python
def sqr(n):
    return np.sqrt(n)

fun = FunctionTransformer(func=sqr)
X['square']= fun.fit_transform(X['Population'])

#COmparision
plt.subplot(2,1,1)
plt.title("Original")
X['Population'].plot(kind='kde')
plt.show()

plt.subplot(2,1,2)
plt.title("Transformed")
X['sqrt'].plot(kind='kde')
plt.show()
```

```python
def square(n):
    return np.power(n,2)

fun = FunctionTransformer(func=square)
X['square']= fun.fit_transform(X['Population'])

#COmparision
plt.subplot(2,1,1)
plt.title("Original")
X['Population'].plot(kind='kde')
plt.show()

plt.subplot(2,1,2)
plt.title("Transformed")
X['square'].plot(kind='kde')
plt.show()
```

```python
def recip(n):
    return np.reciprocal(n)

fun = FunctionTransformer(func=recip)
X['reciprocal']= fun.fit_transform(X['Population'])

#COmparision
plt.subplot(2,1,1)
plt.title("Original")
X['Population'].plot(kind='kde')
plt.show()

plt.subplot(2,1,2)
plt.title("Transformed")
X['reciprocal'].plot(kind='kde')
plt.show()
```

```python
def log(n):
    return np.log(n)

fun = FunctionTransformer(func=log)
X['log']= fun.fit_transform(X['Population'])

#COmparision
plt.subplot(2,1,1)
plt.title("Original")
X['Population'].plot(kind='kde')
plt.show()

plt.subplot(2,1,2)
plt.title("Transformed")
X['log'].plot(kind='kde')
plt.show()
```
