**Missing values Treatment**

Imputing / droping

**Univariate Imputation**

statistical / fillna / constant value / deck imputation

**SimpleImpiter from sklearn**

```python
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer , KNNImputer , IterativeImputer
```

```python
simple = SimpleImputer(missing_values=np.nan , strategy='mean')
```

```python
df[['LoanAmount']] = simple.fit_transform(df[['LoanAmount']])
```

```python
simple = SimpleImputer(missing_values=np.nan , strategy='median')
```

```python
df[['Credit_History']]= simple.fit_transform(df[['Credit_History']])
```

```python
simple = SimpleImputer(missing_values=np.nan , strategy='most_frequent')
```

```python
df[['Self_Employed']] = simple.fit_transform(df[['Self_Employed']])
```

```python
df[['Gender']].isna().sum()
```

**KNNeighbour Imputer**

```python
knn = KNNImputer(missing_values=np.nan , n_neighbors=5)
```

```python
df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term']] = knn.fit_transform(df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term']])
```

```python
knn = KNNImputer(missing_values=np.nan , n_neighbors=10)
```

```python
df[['Loan_Amount_Term']] = knn.fit_transform(df[['Loan_Amount_Term']])
```

```python
df.isna().sum()
```

```python
knn.fit_transform()
```

```python
df.info()
```

```python
df.isna().sum()
```

**IterativeImputer**

```python
iter = IterativeImputer()
```

```python
df[['Credit_History' , 'LoanAmount']] = iter.fit_transform(X = df[['Credit_History' , 'LoanAmount']])
```
