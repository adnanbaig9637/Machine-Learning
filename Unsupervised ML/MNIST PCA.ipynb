```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
df = pd.read_csv(r"C:\Users\User\#DataScience-IRL-Hyd\AI-ML-Inn\Unsupervised ML\mnist_train.csv")
df
```

```python
#Splitting
X = df.drop('label' , axis=1)
y = df['label']

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=23)
```

```python
#Model Before PCA
knn = KNeighborsClassifier()
knn.fit(X_train , y_train)
y_pred = knn.predict(X_test)

metrics.accuracy_score(y_test,y_pred)
```

**Applying PCA : reducing 785 columns in some columns**

```python
#Reducing 785 to 10 colms
pca = PCA(n_components=10)
pca_X_train = pd.DataFrame(pca.fit_transform(X_train))
pca_X_test = pd.DataFrame(pca.fit_transform(X_test))
```

```python
#Model After PCA
%%time
knn = KNeighborsClassifier()
knn.fit(pca_X_train , y_train)
y_pred = knn.predict(pca_X_test)

metrics.accuracy_score(y_test,y_pred)
```

```python
#Reducing 785 to 20 colms
pca = PCA(n_components=10)
pca_X_train = pd.DataFrame(pca.fit_transform(X_train))
pca_X_test = pd.DataFrame(pca.fit_transform(X_test))
```

```python
#Model After PCA
%time
knn = KNeighborsClassifier()
knn.fit(pca_X_train , y_train)
y_pred = knn.predict(pca_X_test)

metrics.accuracy_score(y_test,y_pred)
```

```python
#Finding Optimal score on 2-30 componenets

for i in range(2,30):
    pca = PCA(n_components=i)
    pca_X_train = pd.DataFrame(pca.fit_transform(X_train))
    pca_X_test = pd.DataFrame(pca.fit_transform(X_test))
    
    knn = KNeighborsClassifier()
    knn.fit(pca_X_train , y_train)
    y_pred = knn.predict(pca_X_test)

    print(f"{i} : {metrics.accuracy_score(y_test,y_pred)}")
```

```python
#9 Component giving best score
pca = PCA(n_components=9)
pca_X_train = pd.DataFrame(pca.fit_transform(X_train))
pca_X_test = pd.DataFrame(pca.fit_transform(X_test))

#Model After PCA
knn = KNeighborsClassifier()
knn.fit(pca_X_train , y_train)
y_pred = knn.predict(pca_X_test)

metrics.accuracy_score(y_test,y_pred)
```
