```python
import pandas as pd
import numpy as np
# Define the number of samples
n_samples = 1000
# Generate random data for each column
gender = np.random.binomial(1, 0.5, n_samples) # 0 = Male, 1 = Female
age = np.random.randint(18, 65, n_samples)
openes = np.random.uniform(1, 10, n_samples)
neuroticism = np.random.uniform(1, 10, n_samples)
conscientiousnes = np.random.uniform(1, 10, n_samples)
agreablenes = np.random.uniform(1, 10, n_samples)
extraversion = np.random.uniform(1, 10, n_samples)
# Define the personality labels based on the input values
def get_personality_label(openes, neuroticism, conscientiousnes, agreablenes, extraversion):
     if extraversion > 7:
        return 'extraverted'
     elif neuroticism > 7:
        return 'serious'
     elif conscientiousnes > 7:
        return 'responsible'
     elif agreablenes > 7:
        return 'lively'
     else:
        return 'dependable'
personality_labels = [get_personality_label(o, n, c, a, e) for o, n, c, a, e in zip(openes, neuroticism, conscientiousnes, agreablenes, extraversion)]
# Create the DataFrame
df = pd.DataFrame({
 'Gender': gender,
 'Age': age,
 'Openes': openes,
 'Neuroticism': neuroticism,
 'Conscientiousnes': conscientiousnes,
 'Agreablenes': agreablenes,
 'Extraversion': extraversion,
 'Personality': personality_labels
})
# Print the DataFrame
df
```

```python
import pandas as pd
import numpy as np
# Define the number of samples
n_samples = 1520
# Generate random data for each column
gender = np.random.binomial(1, 0.5, n_samples) # 0 = Male, 1 = Female
age = np.random.randint(18, 35, n_samples)
openes = np.random.randint(1, 10, n_samples)
neuroticism = np.random.randint(1, 10, n_samples)
conscientiousnes = np.random.randint(1, 10, n_samples)
agreablenes = np.random.randint(1, 10, n_samples)
extraversion = np.random.randint(1, 10, n_samples)
self_efficacy = np.random.randint(1, 10, n_samples)
self_esteem = np.random.randint(1, 10, n_samples)
social_monitoring = np.random.randint(1, 10, n_samples)
proactive_personality = np.random.randint(1, 10, n_samples)

# Define the personality labels based on the input values
def get_personality_label(openes, neuroticism, conscientiousnes, agreablenes, extraversion):
     if extraversion > 7:
        return 'extraverted'
     elif neuroticism > 7:
        return 'serious'
     elif conscientiousnes > 7:
        return 'responsible'
     elif agreablenes > 7:
        return 'lively'
     else:
        return 'dependable'
personality_labels = [get_personality_label(o, n, c, a, e) for o, n, c, a, e in zip(openes, neuroticism, conscientiousnes, agreablenes, extraversion)]
# Create the DataFrame
df = pd.DataFrame({
 'Gender': gender,
 'Age': age,
 'Openes': openes,
 'Neuroticism': neuroticism,
 'Conscientiousnes': conscientiousnes,
 'Agreablenes': agreablenes,
 'Extraversion': extraversion,
 'Personality': personality_labels
})
# Print the DataFrame
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop('Personality' , axis=1)
y = df['Personality']

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=23)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train , y_train)

y_pred = knn.predict(X_test)

accuracy_score(y_test , y_pred)
```
