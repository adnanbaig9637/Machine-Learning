```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
X ,y = make_blobs(centers=4 , random_state=23 ,cluster_std=0.6)
#y= clusters name
```

```python
sns.scatterplot(x  = X[:,0] , y = X[:,1] , hue=y)
#we have 4 clusters
```

```python
kmean = KMeans(n_clusters=4, init= 'random' , max_iter=300 , verbose=2 , random_state=23)
kmean.fit(X , y)
```

```python
pred = kmean.predict(X)
```

```python
centroid = kmean.cluster_centers_
#Centers
```

```python
sns.scatterplot(x = X[:,0] , y = X[:,1] , hue=pred)
```

```python
kmean = KMeans(n_clusters=3, init= 'random' , max_iter=300 , random_state=23)
kmean.fit(X , y)
pred = kmean.predict(X)
sns.scatterplot(x = X[:,0] , y = X[:,1] , hue=pred)

#3 cluster
```

```python
kmean = KMeans(n_clusters=2, init= 'random' , max_iter=300 , random_state=23)
kmean.fit(X , y)
pred = kmean.predict(X)
sns.scatterplot(x = X[:,0] , y = X[:,1] , hue=pred)

#2 cluster
```

```python
#Centroid

kmean = KMeans(n_clusters=3, init= 'random' , max_iter=300 , random_state=23)
kmean.fit(X , y)
pred = kmean.predict(X)
centroid = kmean.cluster_centers_




sns.scatterplot(x = X[:,0] , y = X[:,1] , hue=pred)
sns.scatterplot(x = centroid[:,0] , y = centroid[:,1] , c = 'red')
```
